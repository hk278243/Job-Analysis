# Data Cleaning

```python
df.head()
```

```python
for i in df['reviews']:
    print(i)
```

```python
#missing value

df.isnull().sum()
```

```python
df['rating'] = np.where(np.isnan,-999,df['rating'])
df['reviews'] = np.where(np.isnan,-999,df['reviews'])
```

```python
df.isnull().sum()
```

```python
df = df.dropna(axis=0)
df.shape
```

```python
df.isnull().sum()
```

```python
df['rating'] = df['rating'].fillna(-999)
df['reviews'] = df['reviews'].fillna(-999)
```

```python
df.head()
```

```python
# Revies Column
#int reviews, -999

df['reviews_int'] = df['reviews'].apply(lambda x:int(str(x).split()[0]) if len(str(x).split()) >1 else -999)
```

```python
df.head()
```

```python
df['posted_on'].unique()
```

```python
df['posted_on_int'] = df['posted_on'].apply(lambda x:int(str(x).split()[0]) if str(x).split()[0].isnumeric() and str(x).split()[1] in ['Days','Day'] else 0)
```

```python
### How many jobs are there for Data Science profiles

* jobs which has requirement for Machine Learning('Machine')
```

```python
machine_learning_jobs = [i for i in df['job_role'] if "machine learning" in i.lower()]
ba_jobs = [i for i in df['job_role'] if "business analyst" in i.lower()]
```

```python
print('Currently there are {} jobs available in Machine Learning'.format(df[df['job_role'].isin(machine_learning_jobs)].shape[0]))
print('Currently there are {} jobs available in Business Analyst'.format(df[df['job_role'].isin(ba_jobs)].shape[0]))
```
